```python
import hashlib
```

```python
# example with lib
name = '홍길동'
phone_number = '01012345678'
birth_day = '20230101'
salt = 'abc123'

combkey1 = hashlib.sha256((name + phone_number + birth_day + salt).encode()).hexdigest()
print(combkey1)
```

```python
# example with web service
tmp_str = name + phone_number + birth_day + salt
tmp_str
```

```python
import pandas as pd
```

```python
df = pd.read_csv('pidicon2023_dataset.csv', low_memory=False)
df
```

```python
df['combkey'] = df.apply(lambda x: hashlib.sha256((str(x['성별']) +
                                   str(x['생년월일']) +
                                   str(x['가구특성']) +
                                   str(x['홈쇼핑(지수)'])).encode()).hexdigest(), axis=1)
df
```
