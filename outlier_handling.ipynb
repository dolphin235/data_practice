```python
# 시그마 (2-3)
def get_outliers_with_sigma(df, column, size_of_sigma):
    lowest = df[column].mean() - (df[column].std()*size_of_sigma)
    highest = df[column].mean() - (df[column].std()*size_of_sigma)

    outlier_idx = df[column][(df[column] < lowest) | (df[column] > highest)].index
    return outlier_idx
```

```python
col = "소액결제 사용 금액"
outlier_idx = get_outliers_with_sigma(df=df, column=col, size_of_sigma=3)
outlier_idx
```

```python
# 이상치 데이터 확인
df[col].loc[outlier_idx]
```

```python
# 이상치 데이터 행 삭제
df.drop(outlier_idx, axis=0, inplace=True)
df.shape
```

```python
df = pd.read_csv('pidicon2023_dataset.csv', low_memory=False)
df.shape
```

```python
# IQR (1.5) = 약 2.7 시그마
def get_outlier_with_iqr(df, column):
    weight=1.5

    quantile_25 = np.percentile(df[column].values, 25)
    quantile_75 = np.percentile(df[column].values, 75)

    IQR = quantile_75 - quantile_25
    IQR_weight = IQR*weight

    lowest = quantile_25 - IQR_weight
    highest = quantile_75 + IQR_weight

    outlier_idx = df[column][(df[column] < lowest) | (df[column] > highest)].index
    return outlier_idx
```

```python
col = "소액결제 사용 금액"
outlier_idx = get_outlier_with_iqr(df, col)
outlier_idx
```

```python
# 이상치 데이터 확인
df[col].loc[outlier_idx]
```

```python
# 이상치 데이터 행 삭제
df.drop(outlier_idx, axis=0, inplace=True)
df.shape
```
