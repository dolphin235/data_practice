```python
import pandas as pd
```

```python
a = {'A' : [1, 2, 3, 4, 5, 6], 'B' : [10, 20, 30, 40, 50, 60], 'C' : [100, 200, 300, 400, 500, 600]}
df = pd.DataFrame(a)
df
```

## 값 1개 변경

```python
df.loc[2, 'A'] = 3000
df
```

## 특정 열 통째로 변경

```python
df['A'] = [1000, 2000, 3000, 4000, 5000, 6000]
df
```

## 특정 행 통째로 변경

```python
df.loc[2] = [0, 0, 0]
df
```

## 여러 행, 여러 열 동시에 변경

```python
df.loc[[2, 4], ['A', 'C']] = 9999
df
```

## 조건에 해당하는 값 변경

```python
df.loc[df['A'] == 9999, 'B'] = -500
df
```

## replace 이용 특정 값 대체

```python
df = df.replace(-500, 0)
# df = df.replace({'B' : -500}, 0) : 바꿀 값의 범위를 B열로만 한정
df
```
